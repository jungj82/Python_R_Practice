Loading pycaret.

```python
from pycaret.classification import *
setup_survived = setup(train, target = 'Survived')
```

Comparing models.

```python
top5_survived = compare_models(sort='Accuracy', n_select=5)
```

Turning the selected top 5 models.

```python
tune5_survived = [tune_model(i) for i in top5_survived]
```

Blending 5 models into a single model.

```python
blend_survived = blend_models(estimator_list=tune5_survived)
```

Creating a final model based on the entire data

```python
final_survived = finalize_model(blend_survived)
```

Calling in Test.csv to make predictions based on final_survived data

```python
test = pd.read_csv('test.csv')
prediction = predict_model(final_survived, data=test)
```

```python
prediction
```

Column "Label" shows whether the passenger is predicted to survived(1).
The predicted dataframe will be exported to be compared with Kaggle data.

```python
prediction = prediction.rename(columns = {"Label" : "Survived"})
prediction.to_csv('prediction.csv', columns = ['PassengerId','Survived'], index = False)
```
